```python
concrete_data = pd.read_csv('https://cocl.us/concrete_data')
concrete_data.head()
```

## 1.1: Build a baseline model

```python
concrete_data_cols = concrete_data.columns

predictors = concrete_data[concrete_data_cols[concrete_data_cols != 'Strength']]
target = concrete_data['Strength']
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

## 1.2 Build Model:
    - One hidden layer of 10 nodes, and a ReLU activation function
    - Use the adam optimizer and the mean squared error as the loss function.

```python
def regression_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

model = regression_model()
```

### 1.2.1 Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(predictors, target, 
                                                    test_size=0.3, random_state=0)
```

### 1.2.2 Train the model on the training data using 50 epochs.

```python
model.fit(X_train, y_train, epochs=50, verbose=2)
```

### 1.2.3 Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength.

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)
mean_squared_error(y_pred, y_test)
```

### 1.2.4 Repeat steps 1-3 for 50 times

```python
mse = []

for i in range(0, 50):
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, 
                                                    test_size=0.3, random_state=0)
    model.fit(X_train, y_train, epochs=50, verbose=2)
    y_pred = model.predict(X_test)
    mse.append(mean_squared_error(y_pred, y_test))
```

```python
import statistics
print("Mean of the MSE is: ", statistics.mean(mse))
print("Std of the MSE is: ", statistics.stdev(mse))
```

## 2.1 Normalize the data

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
model = regression_model()

mse_norm = []

for i in range(0, 50):
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, 
                                                    test_size=0.3, random_state=0)
    model.fit(X_train, y_train, epochs=50, verbose=2)
    y_pred = model.predict(X_test)
    mse_norm.append(mean_squared_error(y_pred, y_test))
```

```python
print("Mean of the MSE is: ", statistics.mean(mse_norm))
print("Std of the MSE is: ", statistics.stdev(mse_norm))
```

## 3.1 Increase the number of epochs to 100

```python
model = regression_model()

mse_norm_100 = []

for i in range(0, 50):
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, 
                                                    test_size=0.3, random_state=0)
    model.fit(X_train, y_train, epochs=100, verbose=2)
    y_pred = model.predict(X_test)
    mse_norm_100.append(mean_squared_error(y_pred, y_test))
```

```python
print("Mean of the MSE is: ", statistics.mean(mse_norm_100))
print("Std of the MSE is: ", statistics.stdev(mse_norm_100))
```

## 4.1 Increase the number of hidden layers to 3 hidden layers w/ 10 nodes

```python
def regression_model_multiple_layer():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = regression_model_multiple_layer()

mse_norm_multiple_layer = []

for i in range(0, 50):
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, 
                                                    test_size=0.3, random_state=0)
    model.fit(X_train, y_train, epochs=100, verbose=2)
    y_pred = model.predict(X_test)
    mse_norm_multiple_layer.append(mean_squared_error(y_pred, y_test))
```

```python
print("Mean of the MSE is: ", statistics.mean(mse_norm_multiple_layer))
print("Std of the MSE is: ", statistics.stdev(mse_norm_multiple_layer))
```
